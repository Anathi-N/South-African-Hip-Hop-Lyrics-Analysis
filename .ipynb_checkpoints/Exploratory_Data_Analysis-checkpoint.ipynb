```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#open the pickled document term matrix
data = pd.read_pickle('dtm.pkl')
data= data.transpose()
data.head()
```

### Explore the most common words

```python
#Find the top 10 words said by each rapper
top_dict ={}
for c in data.columns:
    top = data[c].sort_values(ascending=False).head(15)
    top_dict[c]= list(zip(top.index, top.values))
    
top_dict
```

```python
#Top 15 words for each rapper
for artist_name, top_words in top_dict.items():
    print(artist_name)
    print(','.join([word for word,count in top_words[0:14]]))
    print('---')
```

```python
#Analyze these top words and add the most common across rappers to stop_words list
from collections import Counter

#pulling out the top 30 words

words =[]
for artist_name in data.columns:
    top= [word for (word, count)in top_dict[artist_name]]
    for t in top:
        words.append(t)
words
```

```python
#Aggregating the list and identifying the most common and the routines
Counter(words).most_common()
```

Observing this only the word 'like may need to be added to stop words

```python
#If over half of the rappers have the word as their top words then we add them to stopwords
add_stop_words =[word for word, count in Counter(words).most_common() if count >= 2  ]
add_stop_words
```

```python
#Updating the document term matrix with the new list of stop words removed
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer

data_clean = pd.read_pickle('data_clean.pkl')

#adding the new stopwords
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)

#recreatint the document term matrix
cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(data_clean.lyrics)
data_stop = pd.DataFrame(data_cv.toarray(), columns= cv.get_feature_names())
data_stop.index = data_clean.index

#pickel this for later use
import pickle
pickle.dump(cv, open('cv_stop.pkl', 'wb'))
data_stop.to_pickle('dtm_stop.pkl')
```

```python
#creating a word cloud for top words by rapper
from wordcloud import WordCloud

wc = WordCloud(stopwords=stop_words, background_color='white', colormap='Dark2',max_font_size= 150, random_state=42)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']= [16,6]

full_names =['A_Reece','Nasty C',  'Stogie T', 'Yugen Blakrok']
#Creating subplots for each comedian
for index, artist_name in enumerate(data.columns):
    wc.generate(data_clean.lyrics[artist_name])
    
    plt.subplot(2,2, index+1)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(full_names[index])
    
plt.show();
```

### Vocabulary Analysis

```python
data
```

```python
artist_name=['a_reece','nasty_c','yugen_blakrok', 'stogie_t']
```

```python
#Find the number of unique words that each comedian uses
#by identifying the none zero items in the document term matrix
unique_list = []
for artist_name in data.columns:
    uniques = data[artist_name].to_numpy().nonzero()[0].size
    unique_list.append(uniques)

#create a new dataframe that contains this unique word count
data_words = pd.DataFrame(list(zip(artist_name,unique_list)), columns=['rapper', 'unique_words'])
data_uniquewords_sort =data_words.sort_values(by ='unique_words')
data_uniquewords_sort
```

```python
#fixing the rapper column

data_words.loc[data_words['unique_words']==186, ['rapper']]= 'A-Reece'
data_words.loc[data_words['unique_words']==206, ['rapper']]= 'Nasty C'
data_words.loc[data_words['unique_words']==232, ['rapper']]= 'Stogie T'
data_words.loc[data_words['unique_words']==229, ['rapper']]= 'Yugen Blakrok'
```

```python
data_words.head()
```

Here is a representation with the full names and unique words

```python
#creating total words
total_list =[]
for rapper in data.columns:
    totals = sum(data[rapper])
    total_list.append(totals)
total_list

#song times from genius.com
track_time =[4.50, 5.33, 4.31, 3.40]
```

```python
#adding new variables based on feature engineering
data_words['total_words']= total_list
data_words['track_time']= track_time
data_words['uniqueness'] =(data_words['unique_words'])/(data_words['total_words'])
data_words['words_per_minute']= (data_words['total_words'])/(data_words['track_time'])

data_words
```

```python
#Plotting these findings
import numpy as np

#creating a scatter plot for the findings
plt. rcParams['figure.figsize']= 10,8
for a, rapper in enumerate(data_words.index):
    x= data_words.unique_words.loc[rapper]
    y= data_words.total_words.loc[rapper]
    plt.scatter(x,y, color='red')
    plt.text(x+1.5, y+0.5, full_names[a], fontsize=10)
    plt.xlim(-0.5, 300)

plt.title('', fontsize=20)
plt.xlabel('Number of Unique Words', fontsize=15)
plt.ylabel('Total Number of Words', fontsize=15)

plt.show();
```

The graph shows that A-Reece and Nasty C are similar although they say a lot they might be repetitive. Yugen Blakrok and Stogie T are similar, they talk less but vary what they say.
